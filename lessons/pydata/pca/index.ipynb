# Analýza hlavních komponent (PCA)

...a jiné metody průzkumu dat a redukce počtu dimenzí

## Co se naučíte

Analýza hlavních komponent (anglicky “principal component analysis” neboli PCA) je metoda na snížení počtu dimenzí mnohorozměrných dat. To se může hodit z hlediska vizualizace nebo kvůli úspoře času/paměti při strojovém učení.

Ještě než začneme, stáhněte si prosím [tento notebook](http://github.com/simecek/naucse.python.cz/blob/praha-jaro2020/lessons/pydata/pca/index.ipynb?raw=true) na svůj počítač a spuštěním následující buňky nahrajte potřebné knihovny.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

%matplotlib inline
sns.set_style('whitegrid')
```

## Jak porovnat víno babičce ve spíži?

Data níže obsahují 13 různých charakteristik 178 vín. Představte si, že chcete tato vína vykreslit do grafu. Případně si představte, že vás babička poprosila, abyste je vyrovnali do polic tak, aby "podobná" vína byla u sebe?

```python
from sklearn.datasets import load_wine
wine = load_wine()

wine_df = pd.DataFrame(wine.data, columns=wine.feature_names)
print(wine_df.shape)
wine_df.head()
```

Jak to uděláte? Jedna možnost je vybrat si dvojici dvou vlastností, která se zásadní pro vaše data. Ale co když se ve víně nevyznáte? Můžete pochopitelně vykreslit grafy všech možných dvojic. 

```python
sns.pairplot(wine_df);
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(wine_df.corr(), annot=True); 
```

Jak vidíme, mnoho charakteristik je korelovaných a zachycuje společné faktory. PCA umožňuje zjednodušit na několik vlastností, které a) jsou nezávislé b) zachycují maximum informace (variability) v datech.

Protože grafy a stěny ve spíži obvykle bývají dvourozměrné, budeme obvykle zjednodušovat z X charakteristik na dvě, abychom tyto potom mohli zakreslit do grafu. Podívejme se, jak to dopadne pro vína.

```python
# Tento blok znormalizuje sloupečky v tabulce, aby měli průměr nula a rozptyl jedna
scaler = StandardScaler()
scaler.fit(wine_df)
wine_df_scaled = scaler.transform(wine_df)
```

```python
# Tento blog spočte PCA souřadnice
pca = PCA(n_components=2)
pca.fit(wine_df_scaled)
wine_pca = pd.DataFrame(pca.transform(wine_df_scaled), columns=['PCA1', 'PCA2'])
wine_pca
```

```python
# A tento je zakreslí do grafu
sns.scatterplot(wine_pca.PCA1, y=wine_pca.PCA2);
```

Splnili jsme úkol, bude babička spokojená? Jsou podobná vína zakreslená poblíž sebe? 

K vínům v databázi lze získat ještě jednu informaci - patří do jedné ze tří kategorií (barva vína?). A skutečně, pokud si body na PCA grafu obarvíme podle kategorie vína, jasně vidíme tři clustery.

```python
sns.scatterplot(wine_pca.PCA1, y=wine_pca.PCA2, hue=wine.target_names[wine.target]);
```

Můžeme se též ptát, kolik procent variability je v prvních dvou PCA komponentách zachyceno.

```python
pca.explained_variance_ratio_
```

## Breast Cancer Wisconsin Dataset

https://www.kaggle.com/faressayah/support-vector-machine-pca-tutorial

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()

col_names = list(cancer.feature_names)
col_names.append('target')
df = pd.DataFrame(np.c_[cancer.data, cancer.target], columns=col_names)
df.head()
```

```python
cancer.target_names, df.shape
```

```python
df.info()
```

```python
sns.pairplot(df, hue='target', vars=['mean radius', 'mean texture', 'mean perimeter', 'mean area', 
                                     'mean smoothness', 'mean compactness', 'mean concavity',
                                     'mean concave points', 'mean symmetry', 'mean fractal dimension'])
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(df.corr(), annot=True) 
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df)
```

```python
scaler = StandardScaler()
scaler.fit(df.drop('target', axis=1))
scaled_data = scaler.transform(df.drop('target', axis=1))
```

```python
pd.DataFrame(scaled_data).head()
```

```python
pca = PCA(n_components=2)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
pd.DataFrame(x_pca)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second Principal Component')
```

## Digits

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 
digits = load_digits()

print(digits.data.shape)

plt.gray() 
plt.matshow(digits.images[1]);
```

```python
plt.matshow(digits.images[1].reshape(1,64)) 
```

```python
digits.images[1].reshape(1,64)
```

```python
digits.images.reshape(1797,64)[1,:]
```

```python
pca = PCA(n_components=2)
pca.fit(digits.images.reshape(1797,64))
x_pca = pd.DataFrame(pca.transform(digits.images.reshape(1797,64)), columns=['PCA1', 'PCA2'])

plt.figure(figsize=(10,10)) 
sns.scatterplot(x_pca.PCA1, y=x_pca.PCA2, hue=digits.target_names[digits.target], legend='full',
                palette=sns.color_palette());
```

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, random_state=0)

x_tsne = pd.DataFrame(tsne.fit_transform(digits.images.reshape(1797,64)), columns=['TSNE1', 'TSNE2'])
plt.figure(figsize=(10,10)) 
sns.scatterplot(x_tsne.TSNE1, y=x_tsne.TSNE2, hue=digits.target_names[digits.target], legend='full',
                palette=sns.color_palette());
```
